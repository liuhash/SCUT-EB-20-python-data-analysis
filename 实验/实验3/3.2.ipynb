```python
import pandas as pd
import numpy as np
df1=pd.read_excel("知识产权年鉴数据.xlsx")
df1=df1.iloc[:,[0,1,4,6,10,17]]
df1.columns=['年份','地区','结案率','专利_申请量','专利_授权量','职务专利授权量百分比']
df1
```

```python
df1['年份'].fillna(method="ffill",inplace=True)
df1['年份']=df1['年份'].astype(int)
df1
```

```python
def process(x):
    if np.isinf(x):
        return 0
    if x>1:
        return 1

df1['结案率'].map(process)
df1.set_index('年份',inplace=True)
df1
```

```python
df2=pd.read_excel("各省人口数（2009-2018）.xlsx")
df2=pd.melt(df2, id_vars =['地区'], value_vars =[i for i in range(2009,2019)],var_name="年份",value_name="人口数")
df2['人口数']=df2['人口数'].map(lambda x:x*10000)
df2.set_index("年份",inplace=True)
df2
```

```python
# print(df1.iloc[31:,:])
# print(df2.iloc[:,-1:])
df=pd.concat([df1.iloc[31:,:],df2.iloc[:,-1:]],axis=1)
df
```

```python
df_gdp=pd.read_excel("gdp/gdp.xlsx")

df_gdp=pd.melt(df_gdp, id_vars =['地区'], value_vars =[i for i in range(2009,2019)],var_name="年份",value_name="GDP")
# 一亿应是10的8次方，但word中操作给的是e+13,那就是10的9次方。
df_gdp["GDP"]=df_gdp["GDP"].map(lambda x:x*10**9)
df_gdp.set_index("年份",inplace=True)
df_gdp
```

```python
df_total_export_import=pd.read_excel("各省进出口总额.xlsx",usecols=[i for i in range(11)])
df_total_export_import.columns=["地区",*[i for i in range(2009,2019)]]
df_total_export_import=pd.melt(df_total_export_import, id_vars =['地区'], value_vars =[i for i in range(2009,2019)],var_name="年份",value_name="进出口总额")
df_total_export_import["进出口总额"]=df_total_export_import["进出口总额"].map(lambda x:x*10**9*6.5286)
df_total_export_import.set_index("年份",inplace=True)
df_total_export_import
```

```python
df_foreign_direct_investment=pd.read_excel("外商直接投资（FDI）.xlsx",usecols=[i for i in range(11)])
df_foreign_direct_investment.columns=["地区",*[i for i in range(2009,2019)]]
df_foreign_direct_investment=pd.melt(df_foreign_direct_investment, id_vars =['地区'], value_vars =[i for i in range(2009,2019)],var_name="年份",value_name="外商直接投资")
df_foreign_direct_investment["外商直接投资"]=df_foreign_direct_investment["外商直接投资"].map(lambda x:x*10**4*6.5286)
df_foreign_direct_investment.set_index("年份",inplace=True)
df_foreign_direct_investment
```

```python
df_social_fixed_capital_in=pd.read_excel("全社会固定资产投入.xlsx",usecols=[i for i in range(11)])
df_social_fixed_capital_in.columns=["地区",*[i for i in range(2009,2019)]]
df_social_fixed_capital_in=pd.melt(df_social_fixed_capital_in, id_vars =['地区'], value_vars =[i for i in range(2009,2019)],var_name="年份",value_name="全社会固定资产投入")
df_social_fixed_capital_in["全社会固定资产投入"]=df_social_fixed_capital_in["全社会固定资产投入"].map(lambda x:x*10**9)
df_social_fixed_capital_in.set_index("年份",inplace=True)
df_social_fixed_capital_in
```

```python
df=pd.concat([df,df_gdp.iloc[:,-1],df_total_export_import.iloc[:,-1],df_social_fixed_capital_in.iloc[:,-1],df_foreign_direct_investment.iloc[:,-1]],axis=1)
df
```

```python
df['人均专利申请量']=df['专利_申请量']/df['人口数']
df
```

```python
# 清洗后的结案率，与word上数据较接近。
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
pca = PCA(n_components=3)
scaler=StandardScaler()
# df.loc[:,['结案率','职务专利授权量百分比','人均专利申请量']]
df_train=scaler.fit_transform(df.loc[:,['结案率','职务专利授权量百分比','人均专利申请量']])
# df_train
pca.fit(df_train)
weights1=pca.explained_variance_ratio_
```

```python
# 未清洗结案率版本,仍未得到word上的数值
pd.options.mode.use_inf_as_na = True
df_jiean_lian=pd.read_excel("知识产权年鉴数据.xlsx")
df_jiean_lian=df_jiean_lian.iloc[31:,[2,3]]
df_jiean_lian.reset_index(drop=True,inplace=True)
df_jiean_lian['立案']=df_jiean_lian['立案'].astype(int)
df_jiean_lian['结案']=df_jiean_lian['结案'].astype(int)
df_jiean_lian['结案率']=df_jiean_lian['结案']/df_jiean_lian['立案']
# df_jiean_lian
df_backup=df.loc[:,['职务专利授权量百分比','人均专利申请量']]
df_backup.insert(0,'结案率',df_jiean_lian['结案率'].values)
df_backup['结案率'].fillna(0,inplace=True)
df_backup['结案率'].replace([np.inf], 0)
any(np.isinf(df_backup['结案率']))
# df_backup
df_backup_train=scaler.fit_transform(df_backup)
# df_backup_train
pca = PCA()
pca.fit(df_backup_train)
pca.explained_variance_ratio_
```

```python
df_sum_temp=weights1*df.loc[:,['结案率','职务专利授权量百分比','人均专利申请量']]
df['知识产权保护系数']=df_sum_temp.sum(axis=1)*4.86
df
```

```python
df['人均GDP']=df['GDP']/df['人口数']
df
```

```python
df['经济开放度']=df['进出口总额']/df['GDP']
df
```

```python
middle_gdp=df['GDP'].median()
# middle_gdp
df['GDP哑变量']=df['GDP'].apply(lambda x:1 if x>middle_gdp else 0)
df
```

```python
# df.drop('人均专利授权量',axis=1,inplace=True)
df.insert(10,'人均专利授权量',df['专利_授权量']/df['人口数'])
df
```
