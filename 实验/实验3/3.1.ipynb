```python
import pandas as pd
df1=pd.read_excel("GDP/2009-2013地区生产总值.xls",usecols=[0,2,3,4,5,6],header=5)
df1.columns=['地区',2009,2010,2011,2012,2013]
df1.set_index("地区",inplace=True)
df1
```

```python
path="GDP/{}各地区生产总值.xls"
df2=pd.read_excel(path.format(2014),usecols=[0,2],header=5)
for i in range(2015,2018):
    df2=pd.concat([df2,pd.read_excel(path.format(i),usecols=[2],header=5)],axis=1)
df2.columns=['地区',*[i for i in range(2014,2018)]]
df2.set_index("地区",inplace=True)
df2
```

```python
df3=pd.read_excel(path.format(2018),usecols=[0,2],header=8,skiprows=list(range(40,44)))
df3.columns=["地区",2018]
df3["地区"]=df3["地区"].map(lambda x:x.replace(" ",''))
df3.set_index("地区",inplace=True)
df3
```

```python
# print(df3.iloc[:,[0]])
# print(df3.index==df2.index)
df4=pd.concat([df1,df2,df3],axis=1)
df4.to_excel("GDP/gdp.xlsx")
df4
```
