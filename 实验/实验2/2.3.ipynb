```python
import pandas as pd
import numpy as np
import os
import sys
import re
import keyword
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity =  "all"
```

```python
#1.用pd.read_excel()读取广东知识产权微博中的数据。
data = pd.read_excel('./广东知识产权微博.xls', index_col=0)
data.head()
```

```python
#2．重新构建data1选取需要的列['微博内容', '时间', '点赞数', '评论数', '转发数']
data1 = data[['微博内容', '时间', '点赞数', '评论数', '转发数']]
data1.head()
```

```python
#3．首先对微博内容这一列进行处理，去掉其中一些不需要的字符串
data1.loc[:,'微博内容'] = data1.loc[:,'微博内容'].copy().str.replace('...','',regex=False)
data1.loc[:,'微博内容'] = data1.loc[:,'微博内容'].copy().str.replace('\u200b','',regex=False)
data1.loc[:,'微博内容'] = data1.loc[:,'微博内容'].copy().str.replace('O网页链接','',regex=False)
data1.loc[:,'微博内容'] = data1.loc[:,'微博内容'].copy().str.replace('\u3000','',regex=False)
data1
```

```python
#4.对时间进行处理，只保留年份，用transform（）函数
data1['时间'] = data1['时间'].transform(lambda x: x.split('-')[0])
data1.head()
```

```python
#5.计算这个微博每一年的发文量，在data1中新增发文量这一列，用对应年份的发文量进行填充。
time = data1['时间'].groupby(data1['时间']).count()
time
data1['发文数量'] = data1['时间'].map(time)
data1.head()
```

```python
#6.分析该省微博对知识产权的重视程度，用知识产权平均字数（一年的知识产权类字数/这一年全部文章字数，知识产权类字数是指包含知识产权、专利、商标、版权和著作权5个关键字其中一个或多个的微博的总字数）衡量，在data1中新增知识产权平均字数这一列，用对应年份的知识产权平均字数进行填充。
#统计字数
data1['字数'] = data1['微博内容'].apply(lambda x: len(str(x)))
data1.head()

keywords = ['知识产权','专利','商标','版权','著作权']

def sum_words(content):
    content = str(content)
    if any(key in content for key in keywords ):
        return len(content)
    else:
        return 0 
        

data1['知识产权类字数'] = data1.apply(lambda x: sum_words(x['微博内容']),axis=1)
data1

df_new = data1.groupby(data1["时间"]).sum()
average_words =df_new["知识产权类字数"]/df_new["字数"]
average_words

data1['知识产权平均字数'] = data1['时间'].map(average_words)
data1.drop(columns=['字数','知识产权类字数'],inplace=True)
data1
```

```python
#存储
data1.to_excel("广东知识产权数据.xlsx", index=False)
```
