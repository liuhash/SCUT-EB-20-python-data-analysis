```python
import pandas as pd
import numpy as np
import os
import sys
import re
import keyword
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity =  "all"
```

```python
#1、读取数据：人民日报.xlsx
data = pd.read_excel('./人民日报.xlsx', index_col=0)
data.head()
```

```python
#2、新增年份列
data.insert(loc=0,column='年份',value = data['标题'].str[0:4])
data.head()
```

```python
#3.使用正则表达式去掉文章中多余的空白以及，。？；：‘’、！（）《》等符号
strsub = re.compile(r'[/:*?"\'<>|\\，。？；：‘’、！（）《》 ]')
data['正文'] = data['正文'].astype('str').apply(lambda x: strsub.sub('',x))
data.head()
```

```python
#遍历所有文章的内容，若文章内容中存在知识产权， 专利， 商标， 版权， 著作权五个关键词中的一个或者多个，即记为知识产权类文章。
keyword = ['知识产权','专利','商标','版权','著作权']
data['符合知识产权类文章'] = data['正文'].apply(lambda x: any(key in x for key in keyword ))
data
```

```python
#5.计算所有文章字数
new_data = data.copy()
new_data['字数'] = new_data['正文'].apply(lambda x: len(x))
new_data.head()
```

```python
#计算知识产权类文章字数
def choose(know,sum_words):
    if know: 
        return sum_words
    else: 
        return 0
    
    
new_data['konw_words'] = new_data.apply(lambda x : choose(x['符合知识产权类文章'],x['字数']),axis=1)
new_data
```

```python
def sum_all(df):
    return (df['年份'].count(),df['符合知识产权类文章'].sum(),df['字数'].sum(),df['konw_words'].sum())
    
new_data.groupby(new_data['年份']).apply(sum_all)

df_rates = pd.DataFrame(data={'总发文量':[33233, 19943],'知识产权类文章数量':[1463,957],
                           '全部文章总字数':[37696955,24978182],'知识产权类文章总字数':[3625927,2766185]},
                          index=[2018,2019])
df_rates['知识产权类文章数量占比'] = df_rates['知识产权类文章数量']/df_rates['总发文量']
df_rates['知识产权类文章总字数占比'] = df_rates['知识产权类文章总字数']/df_rates['全部文章总字数']
df_rates['知识产权类文章平均字数'] = df_rates['知识产权类文章总字数']/df_rates['知识产权类文章数量']
df_rates
```
