```python
#进行U-A伪装
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.87 Safari/537.36 SE 2.X MetaSr 1.0',
        'Cookie': 'SINAGLOBAL=1175580988061.2888.1568130821501; _s_tentry=www.baidu.com; Apache=9313484394402.018.1636295461368; UOR=www.baidu.com,weibo.com,www.baidu.com; ULV=1636295461372:1:1:1:9313484394402.018.1636295461368:; WBtopGlobal_register_version=2021110722; XSRF-TOKEN=VMWxEGYlpPl_92Ezn6Vlf1Zf; login_sid_t=e362a6c182743d45451f8c1500d065c5; cross_origin_proto=SSL; appkey=; SSOLoginState=1636964990; SUBP=0033WrSXqPxfM725Ws9jqgMF55529P9D9WWbg_kXyLHk1dF_P78xIpZi5JpX5KMhUgL.FoMNeoz4e0q0eK-2dJLoIEzLxKML1KBLBKnLxKqL1hnLBoMcSK-71KBEeKq7P0q7SK-E; ALF=1668674029; SCF=Aub1Od9YbFKpCeB94cPrtdElGMVJfCZFld2hJFh8lQVpjk8J70NBsLYbzz5olwS8nVYgUd7DNmpc1ZZ8GKdthkQ.; SUB=_2A25MkM4hDeRhGeFJ6VAY8yjPyjmIHXVv57jprDV8PUNbmtAKLXWnkW9NfGCtUAJz3wt0PxlxHyiAwhOE_vUS6tYQ; WBPSESS=WJF1PaMcQNB28UrLifgtWcnX8EvAe38gvHhoV95jIo5wuf_g1VUHCWa8zDViIYmQykJubPLeg0K2ztWJu8-cVckSSWfiEWx_KQI18vpM4Ii4Na-cyEVbRGdyOoAYCOXohZsKXAhF3YkkJCVT4mwXLA=='
        }

#设置微博发布的文本内容、发布时间、来源、转发数、评论数、点赞数、起始页数
text_raw_list = []  
created_at_list = []  
source_list = []  
reposts_count_list = []  
comments_count_list = []  
attitudes_count_list = []  
index = 1 
#爬取页面数据
while True:
    url1 = f'https://weibo.com/ajax/statuses/mymblog?uid=2611704875&page={index}&feature=0'
    respond = requests.get(url1, headers=headers)
    page_json = respond.json()
    #判断是否抓完所有内容，是则停止
    if page_json['data']['list'] == []: 
         break
    # 统计所有内容
    count = len(page_json['data']['list'])
    #遍历所有内容进行数据解析
    for i in range(count):
        created_at_list.append(page_json['data']['list'][i]['created_at'])
        source_list.append(page_json['data']['list'][i]['source'])
        reposts_count_list.append(page_json['data']['list'][i]['reposts_count'])
        comments_count_list.append(page_json['data']['list'][i]['comments_count'])
        attitudes_count_list.append(page_json['data']['list'][i]['attitudes_count'])
        if (page_json['data']['list'][i]['isLongText']==False): # 如果不是长文本（不需要展开）
            text_raw_list.append(page_json['data']['list'][i]['text_raw'])            
        else:
            mblogid = page_json['data']['list'][i]['mblogid']
            url2 = f'https://weibo.com/ajax/statuses/longtext?id={mblogid}'# 获取长文本请求url
            response = requests.get(url2, headers=headers)
            try:
                more_json = response.json()
            except JSONDecodeError:
                print(JSon编码错误)
                continue
            if len(more_json['data'].keys())>0:
                text_raw_list.append(more_json['data']['longTextContent'])
            else:
                text_raw_list.append(page_json['data']['list'][i]['text_raw'])    
        print(f'第{index}页存储完成')
    index += 1
    time.sleep(random.random())
#存储数据
data = {
    '微博内容': text_raw_list,
    '时间': created_at_list,
    '来源': source_list,
    '点赞数': attitudes_count_list,
    '评论数': comments_count_list,
    '转发数': reposts_count_list,
}
df = pd.DataFrame(data)
df.to_excel('北京知识产权官方微博.xls')
```
